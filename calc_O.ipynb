```python
import matplotlib.pyplot as plt
from vec3 import Vec3, norm2
from math import radians, atan, atan2, sin, cos, tan, pi
```

```python
def calc_O(F, B, theta):
    L = F - B
    Ox = norm2(L) / tan(theta)
    phi = atan2(L.y, L.x)
    alpha = phi - pi/2.0
    return Vec3(cos(alpha)*Ox, sin(alpha)*Ox) + B
```

```python
def forceAspect(aspect=1):
    axes = plt.gca()
    extent = {
        'x': list(axes.get_xlim()),
        'y': list(axes.get_ylim())
    }
    axes.set_aspect(abs((extent['x'][1]-extent['x'][0]) / (extent['y'][1]-extent['y'][0])) / aspect)

def plot_point(p:Vec3, color='black'):
    ax = plt.gca()
    plt.plot([p.x], [p.y], 'x', color=color)

def plot_line(p:Vec3, q:Vec3, style):
    ax = plt.gca()
    plt.plot([p.x, q.x], [p.y, q.y], style)

def test_calc_O(F, B, theta):
    O = calc_O(F, B, theta)
    fig, ax = plt.subplots(figsize=[5, 5])
    plot_point(F, 'red')
    plot_point(B, 'blue')
    plot_line(F, B, 'k--')
    plot_point(O)
    pts = [F.x, F.y, B.x, B.y, O.x, O.y]
    max_lim = max(pts)
    min_lim = min(pts)
    margin = 1
    lims = (min_lim - margin, max_lim + margin)
    plt.xlim(lims)
    plt.ylim(lims)
    plt.xlabel('x')
    plt.ylabel('y')
    forceAspect()
```

```python
F = Vec3(-0.5,-2)
B = Vec3(0,1)
theta = radians(-45)
test_calc_O(F, B, theta)
```
