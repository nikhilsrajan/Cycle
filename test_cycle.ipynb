```python
from cycle import Cycle
import matplotlib.pyplot as plt
from math import radians
```

```python
def test_move(cy, steps:int=0, theta:float=0.0, plot:bool=False):
    cy.set_theta(theta)
    step = 1
    while step <= steps:
        cy.move_one()
        step += 1
    if plot:
        Fx, Fy, Bx, By = cy.unpack_path_history()
        fig, ax = plt.subplots(figsize=[10, 10])
        plt.plot(Fx, Fy, 'r--')
        plt.plot(Bx, By, 'b--')
        pts = Fx + Fy + Bx + By
        min_lim = min(pts)
        max_lim = max(pts)
        margin = 1
        lims = (min_lim - margin, max_lim + margin)
        plt.xlim(lims)
        plt.ylim(lims)
        plt.xlabel('x')
        plt.ylabel('y')
        plt.show()
```

```python
cy = Cycle()
test_move(cy, steps=50, theta=radians(0))
test_move(cy, steps=50, theta=radians(10))
test_move(cy, steps=50, theta=radians(15))
test_move(cy, steps=50, theta=radians(20))
test_move(cy, steps=50, theta=radians(25))
test_move(cy, steps=50, theta=radians(30))
test_move(cy, steps=50, theta=radians(35))
test_move(cy, steps=10, theta=radians(20))
test_move(cy, steps=10, theta=radians(10))
test_move(cy, steps=10, theta=radians(0))
test_move(cy, steps=10, theta=radians(-10))
test_move(cy, steps=10, theta=radians(-20))
test_move(cy, steps=50, theta=radians(-30))
test_move(cy, steps=10, theta=radians(-20))
test_move(cy, steps=10, theta=radians(-10))
test_move(cy, steps=50, theta=radians(0))
test_move(cy, plot=True)
```
